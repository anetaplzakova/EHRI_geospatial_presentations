```python
import pandas as pd
import geopandas as gpd
%matplotlib inline 
import matplotlib.pyplot as plt
```

```python
import matplotlib
matplotlib.rcParams.update({'font.size': 20, 'text.color': 'black', 'lines.color': 'black'})
matplotlib.rcParams.keys()
```

### Load borders of historical districts

```python
districts = gpd.read_file("data/MGS-prague_districts_borders-v1.gpkg")
districts = districts.to_crs(("EPSG:4326"))
districts
```

```python
districts.plot(facecolor="none", edgecolor='grey', figsize=(18, 18))
```

### Load dataset with places of residence of Jews in Prague

```python
houses = gpd.read_file("data/MGS-prague_jewish_residency_during_occupation-v1-eng.gpkg")
houses.head()
```

```python
fig, ax = plt.subplots(figsize=(20,14))
districts.plot(facecolor="none", edgecolor='grey', ax=ax)
houses.plot(cmap="OrRd", column="murdered", scheme="Quantiles", ax=ax)
```

### Load Stolpersteine dataset

```python
stolpersteine = gpd.read_file("data/STS-stolpersteine_prague-v2_20230930.gpkg")
stolpersteine = stolpersteine.to_crs("EPSG:4326")
stolpersteine.head()
```

```python
len(stolpersteine)
```

```python
fig, ax = plt.subplots(figsize=(18, 18))
stolpersteine.plot(ax=ax)
districts.plot(facecolor="none", edgecolor='grey', ax=ax)
```

```python
dh = districts.sjoin(houses, how="left")
dh.head()
```

```python
dhgrouped = dh[["quarter_en", "murdered"]].groupby(['quarter_en']).sum(['murdered'])
dhgrouped.plot(kind = "barh", figsize=(10, 10))
```

```python
ds = districts.sjoin(stolpersteine, how="left")
ds.head()
```

```python
dsgrouped = ds[["quarter_en"]].groupby(['quarter_en'])['quarter_en'].count()
dsgrouped.plot(kind="barh", figsize=(10, 10))
```

```python
hs = dhgrouped.join(dsgrouped, on="quarter_en")
hs.rename({'quarter_en': 'stolpersteine'}, axis=1, inplace=True)
hs['proportion'] = hs['stolpersteine'] / hs['murdered']
hs.sort_values(['proportion'], ascending=[False])
```

```python
diststat = districts.merge(hs, on='quarter_en', how='left')
fig, ax = plt.subplots(figsize = (28, 20))
diststat[(diststat["murdered"]>0)].plot(cmap="OrRd", column="proportion", legend=1, ax=ax)
diststat.apply(lambda x: ax.annotate(text=x['quarter_en'] + ': ' + str(x['stolpersteine']) + '/' + str(x['murdered']), xy=x.geometry.centroid.coords[0], ha='center'),axis=1);
districts.plot(facecolor="none", edgecolor='grey', ax=ax)
```

Let's try grid data analysis...

Gridding adapted from: James Brennan, [Fast and easy gridding of point data with geopandas](https://james-brennan.github.io/posts/fast_gridding_geopandas/)

```python
import numpy as np
import shapely

# total area for the grid
xmin, ymin, xmax, ymax= districts.total_bounds
# how many cells across and down
n_cells=80
cell_size = (xmax-xmin)/n_cells
# projection of the grid
crs = "EPSG:4326"
# create the cells in a loop
grid_cells = []
for x0 in np.arange(xmin, xmax+cell_size, cell_size ):
    for y0 in np.arange(ymin, ymax+cell_size, cell_size):
        # bounds
        x1 = x0-cell_size
        y1 = y0+cell_size
        grid_cells.append( shapely.geometry.box(x0, y0, x1, y1) )
cell = gpd.GeoDataFrame(grid_cells, columns=['geometry'], crs=crs)
```

```python
cell = cell.assign(gridid=range(0,len(cell)))
cell
```

```python
gridh = cell.sjoin(houses, how='left')
gridh[gridh.murdered > 10]
```

```python
gridh2 = gridh.groupby('gridid').sum("murdered")
```

```python
gridh3 = cell.join(gridh2, on="gridid", how="inner", rsuffix="_right")
fig, ax = plt.subplots(figsize = (28, 20))
gridh3.plot(figsize = (28, 20), cmap="OrRd", column="murdered", legend=1, ax=ax)
districts.plot(facecolor="none", edgecolor='grey', ax=ax)
```

```python
grids = cell.sjoin(stolpersteine, how='left')
grids
```

```python
grids2 = grids.groupby('gridid').count()
grids2
```

```python
grids3 = cell.join(grids2, on="gridid", how="inner", rsuffix="_right")
fig, ax = plt.subplots(figsize = (28, 20))
grids3.plot(figsize = (28, 20), column="geometry_right", cmap="OrRd", legend=1, ax=ax)
districts.plot(facecolor="none", edgecolor='grey', ax=ax)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(30, 12))

gridh3.plot(figsize = (28, 20), cmap="OrRd", column="murdered", legend=1, ax=axes[0])
grids3.plot(figsize = (28, 20), column="geometry_right", cmap="OrRd", legend=1, ax=axes[1])
```
